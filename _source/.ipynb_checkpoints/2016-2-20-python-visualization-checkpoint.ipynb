```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
from ggplot import *
import pygal
import numpy as np
```

```python
iris = pd.read_csv("https://raw.githubusercontent.com/pydata/pandas/master/pandas/tests/data/iris.csv")
```

```python
iris.plot(kind="scatter", x="SepalLength", y="SepalWidth")
```

```python
sns.color_palette(n_colors=3)
```

```python
sns.jointplot(x="SepalLength", y="SepalWidth", data=iris)
```

```python
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.io import output_notebook
output_notebook()
```

```python
SepalLength = iris["SepalLength"]
SepalWidth = iris["SepalWidth"]

# generate a tooltip when hovering over points
source = ColumnDataSource(data=dict(SepalLength=SepalLength,
                                   SepalWidth=SepalWidth,
                                   Name=iris["Name"]))
hover = HoverTool(tooltips=[("Name", "@Name"),
                           ("(Length, Width)", "(@SepalLength, @SepalWidth)")])

TOOLS=["pan,wheel_zoom,box_select,reset", hover]
p = figure(plot_width=500, plot_height=500, title=None, tools=TOOLS)
p.scatter(x=SepalLength, y=SepalWidth, 
          radius=SepalLength * SepalWidth / 400, source=source)
show(p)
```
